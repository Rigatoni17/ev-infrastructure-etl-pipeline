# SQL Demo (state features)
I loaded state_features_2022 into MySQL (db ev_project, table state_features_2022). This notebook shows how to query it. The load used local infile with the CSV columns reordered correctly.

````python
import pandas as pd
try:
    from sqlalchemy import create_engine
    import pymysql  # noqa
except ModuleNotFoundError:
    raise SystemExit("Install PyMySQL to run this cell: pip3 install pymysql")

password = "YOUR PASSWORD HERE"
engine = create_engine(f"mysql+pymysql://root:{password}@127.0.0.1:3306/ev_project")
query = "SELECT State, chargers_per_100k, stations FROM state_features_2022 ORDER BY chargers_per_100k DESC LIMIT 10"
df = pd.read_sql(query, engine)
print(df)
````
